```python
plt.figure(figsize=(6,4))
sns.countplot(train,x="three_g",hue="touch_screen",color="pink")
plt.title("3G mobile's touch_screen")
plt.xlabel("Three_G Network")
plt.ylabel("Device Count")
plt.show()
```
